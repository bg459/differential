```python
import pandas as pd

df = pd.read_csv("jan21.csv")
df.columns = ['company', 'bought', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'sig_cum']
```

```python
## correlation between buying and each months views

import seaborn as sn
import matplotlib.pyplot as plt

data = pd.DataFrame(df, columns = ['bought', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'sig_cum'])
corrMatrix = data.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```
